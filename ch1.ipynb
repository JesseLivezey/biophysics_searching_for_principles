# Problem 1

## (a)

```python
T = 1.e6 # total time in seconds
dt = .01 # bin size in seconds
rate = 10 # counts per second

n_bins = int(T/dt)
thresh = dt*rate
```

```python
bins = (np.random.rand(n_bins) < thresh).astype(int)
```

```python
taus = range(1, 11)
max_count = 150
hist_counts = np.zeros((len(taus), max_count))
counts = []

for ii in taus:
    bins_per_tau = int(ii/dt)
    taus_per_time = int(T/ii)
    jump = int(n_bins/bins_per_tau)
    count_list = []
    for jj in xrange(taus_per_time):
        count = bins[jj:-1:jump].sum()
        hist_counts[ii-1, count] += 1
        count_list.append(count)
    counts.append(np.array(count_list))
```

```python
for ii in xrange(len(taus)):
    plt.plot(hist_counts[ii]/hist_counts[ii].sum())
    print counts[ii].mean(), counts[ii].mean()/(ii+1), counts[ii].std(), np.sqrt(counts[ii].mean())
plt.show()
```
